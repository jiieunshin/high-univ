### 데이터의 요약

데이터 프레임은 행(row)과 열(column)로 이루어져 있다. 행은 관측치라고도 부르며, 각 관측치의 측정된 정보가 해당 열 위치에 기술되어 있다. 따라서 데이터는 각 열을 기준으로 요약하며, 이 절에서도 column 별 `평균, 표준편차, 최댓값/최솟값, 분위수` 등 요약된 수치들을 출력한다. 이러한 요약된 수치들을 `통계량`이라고도 부른다.

실습할 데이터는 화재출동 데이터(`fire_calling_summary.csv`)이다. 화재출동 데이터는 2017년부터 2021년의 5년간 서울시 5개 구별 (강동구, 강서구, 강남구, 강북구, 금천구) 자료를 담고 있다.

```python
import pandas as pd
```

```python
df = pd.read_csv("data/fire_calling_summary.csv", encoding = 'cp949')
df
```

### 열의 요약 통계

전체 데이터를 요약하는 메소드는 `describe`이다.

```python
df.describe()
```

뭔가 문제가 있다. 무엇일까?

<details>
<summary>접기/펼치기</summary>

> 1. `화재발생연도`는 요약이 필요없다.

> 2. `사망자수`와 `출동횟수`는 불필요한 소숫점으로 과도하게 표현되어있다.

> 3. `재산피해금액`은 숫자가 너무 커서 보기가 힘들다.

</details>

그렇다면 어떻게 고치는 것이 좋을까?

<details>
<summary>접기/펼치기</summary>

> 1. `화재발생연도`: 요약 대상에서 제외한다.

> 2. `사망자수`와 `출동횟수`는 요약 후 결과를 소숫점 둘 째 자리에서 반올림하여 표현한다.

> 3. `재산피해금액`은 만 원 단위로 표현한다.

</details>

먼저 요약한 데이터를 `df_summ`으로 저장하자.

```python
df_summ = df.describe()
```

`사망자수`와 `출동횟수`의 소수점을 조정한다.

```python
df_summ['사망자수'] = df_summ['사망자수'].round(2)
df_summ['출동횟수'] = df_summ['출동횟수'].round(2)
df_summ
```

`재산피해금액`의 단위를 바꿔준다.

```python
df_summ['재산피해금액'] = df_summ['재산피해금액'] / 10000
df_summ
```

`재산피해금액`도 소수점 조정을 한다. 이 경우 소수점 4째 자리에서  하였다.

```python
df_summ['재산피해금액'] = df_summ['재산피해금액'].round(4)
df_summ
```

열 이름을 이해하기 쉽게 단위로 표현해주었다.

```python
df_summ.columns = ['화재발생연도', '사망자수', '재산피해금액(만원)', '출동횟수']
```

```python
df_summ
```

마지막으로 `화재발생연도`를 제거하였다.

```python
df_summ = df_summ.drop(columns = '화재발생연도')
df_summ
```

### 해당 연도 요약 통계

위에서는 전체 데이터를 요약해보았다. 이번에는 범주별로 나누어 `연도` 혹은 `시군구`의 자료를 각각 요약해보자.

2017년 부터 2021년 5년간의 자료가 있기 때문에 2021년 자료만 보고 싶은 경우 다음과 같이 원하는 열을 추출한 뒤 `loc`메소드로 인덱싱해주면 된다.

```python
df_2021 = df.loc[df['화재발생연도']==2021]
df_2021
```

`describe`메소드로 자료를 요약한다. 이 때 표현방식을 바꾸고싶다면 위의 과정을 동일하게 거치면 된다.

```python
df_2021.describe()
```

2021년 중에 출동 횟수가 많은 지역구대로 정렬 해보자.

정렬하기 위해서는 `sort_values()`메소드를 사용한다. 괄호 안에 정렬할 열을 지정한다.

```python
df_2021.sort_values(by="출동횟수")
```

`ascending=False` 선택문을 추가하면 출동 횟수가 많은 순으로 내림차순 정렬을 한다.

```python
df_2021.sort_values(by="출동횟수", ascending=False)
```

### 해당 시군구 통계

이번에는 강남구의 자료를 보자.

```python
df_강남구 = df.loc[df['시군구']=="강남구"]
df_강남구
```

`describe`메소드를 용하면 5년간 강남구의 기초통계량을 출력한다.

```python
df_강남구.describe()
```

5년간 강남구의 `출동횟수`가 많은 순으로 출력해보자.

```python
df_강남구.sort_values(by="출동횟수", ascending=False)
```
